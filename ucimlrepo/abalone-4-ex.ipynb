```python
from ucimlrepo import fetch_ucirepo, list_available_datasets
data = fetch_ucirepo(name='Abalone')
df = pd.DataFrame(data= data.data.original)
```

```python
def traditional_outlier(df,x):
    q1  =  df[x].quantile(.25)
    q3  =  df[x].quantile(.75)
    iqr = q3-q1
    df['Traditional']  = np.where(df[[x]]<(q1-1.5*iqr),-1,
                         np.where(df[[x]]>(q3+1.5*iqr),-1,1))
    return df
```

```python
traditional_outlier(df,'Rings')
```

```python
df = df[df['Traditional'] == 1]
```
